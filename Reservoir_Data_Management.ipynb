```python
import pandas as pd
import numpy as np

# 1. بناء بيانات محاكاة لبئر نفطي (قراءات ضغط وإنتاج)
well_data = {
    'Date': ['2026-06-01', '2026-06-02', '2026-06-03', '2026-06-04', '2026-06-05'],
    'Bottomhole_Pressure (psi)': [3200, 3150, 3110, 3070, 3020],
    'Oil_Production (bbl/day)': [1200, 1180, 1150, 1120, 1090],
    'Water_Cut (%)': [1.2, 1.5, 1.8, 2.2, 2.5]
}

# 2. تحويل القاموس إلى DataFrame ذكي باستخدام Pandas
df = pd.DataFrame(well_data)

# 3. عرض الجدول بشكل منسق واحترافي
df
```

```python
# 1. استخراج أعلى معدل إنتاج نفط سُجل في البئر
max_oil = df['Oil_Production (bbl/day)'].max()

# 2. استخراج أقل ضغط قاع بئر سُجل (Bottomhole Pressure)
min_pressure = df['Bottomhole_Pressure (psi)'].min()

# 3. طباعة النتائج بشكل هندسي منسق
print(f"📊 تحليل بيانات البئر الفوري:")
print(f"-----------------------------------")
print(f"🔹 أقصى معدل إنتاج للنفط: {max_oil} برميل/يوم")
print(f"🔹 أقل ضغط سُجل عند القاع: {min_pressure} psi")
print(f"-----------------------------------\n")

# 4. الأمر السحري الذي يعطيك تقريراً إحصائياً كاملاً عن الجدول (المتوسط، الانحراف المعياري، إلخ)
print("📈 التقرير الإحصائي الشامل للبيانات:")
df.describe()
```

```python
# 1. تحديد الشرط الهندي: نريد الأيام التي يكون فيها Water Cut أكبر من 2%
critical_condition = df['Water_Cut (%)'] > 2.0

# 2. تطبيق الشرط على الجدول لاستخراج الأيام الحرجة فقط
critical_days = df[critical_condition]

# 3. عرض الأيام الحرجة
print("⚠️ الأيام الحرجة التي تجاوزت فيها نسبة الماء 2%:")
critical_days
```

```python
# 1. حساب الإنتاج التراكمي للنفط (Cumulative Production) برميل
df['Cumulative_Oil (bbl)'] = df['Oil_Production (bbl/day)'].cumsum()

# 2. عرض الجدول النهائي بعد إضافة عامود الإنتاج التراكمي
print("✅ الجدول الهندسي المتكامل للبئر بعد حساب الإنتاج التراكمي:")
df
```
