```python
import sqlite3
import pandas as pd

# 1. إنشاء اتصال بقاعدة بيانات في الذاكرة (In-Memory Database)
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# 2. كتابة أمر SQL لإنشاء جدول الإنتاج اليومي
cursor.execute('''
CREATE TABLE IF NOT EXISTS production_data (
    Date TEXT,
    Well_Name TEXT,
    Oil_Prod REAL,
    Water_Cut REAL,
    BHP REAL
)
''')

# 3. بيانات الحقل المراد حقنها في قاعدة البيانات
field_records = [
    ('2026-06-01', 'Well-A', 1200, 1.2, 3200),
    ('2026-06-02', 'Well-A', 1180, 1.5, 3150),
    ('2026-06-03', 'Well-B', 2500, 0.5, 3400),
    ('2026-06-04', 'Well-B', 2450, 0.8, 3350),
    ('2026-06-05', 'Well-A', 1090, 2.5, 3020)
]

# 4. إدخال البيانات باستخدام SQL
cursor.executemany('INSERT INTO production_data VALUES (?, ?, ?, ?, ?)', field_records)
conn.commit()

print("🛢️ قاعدة البيانات النفطية جاهزة، وتم إنشاء جدول الإنتاج بنجاح!")
```

```python
# 1. كتابة أمر استعلام SQL لفلترة الإنتاج العالي
# هنا نطلب جلب كافة الأعمدة للأبيار التي تنتج أكثر من 1100 برميل
sql_query = '''
SELECT * FROM production_data 
WHERE Oil_Prod > 1100
'''

# 2. تنفيذ الاستعلام وجلب البيانات مباشرة إلى Pandas DataFrame
df_high_prod = pd.read_sql_query(sql_query, conn)

# 3. عرض النتيجة المهندسة
print("📊 الآبار ذات الإنتاج العالي المستخرجة عبر SQL:")
print("--------------------------------------------------")
print(df_high_prod)
```

```python
# 1. كتابة أمر SQL لحساب المتوسط وأعلى قيمة لكل بئر
sql_summary = '''
SELECT 
    Well_Name,
    AVG(Oil_Prod) AS Avg_Oil_Production,
    MAX(BHP) AS Max_Bottom_Hole_Pressure
FROM production_data
GROUP BY Well_Name
'''

# 2. تنفيذ الاستعلام وجلب الملخص المكمني إلى Pandas
df_summary = pd.read_sql_query(sql_summary, conn)

# 3. عرض التقرير النهائي للمكامن
print("📊 تقرير أداء الآبار المستخرج عبر الـ SQL Aggregation:")
print("--------------------------------------------------")
print(df_summary)

# 4. إغلاق الاتصال بقاعدة البيانات بشكل آمن ونظيف
conn.close()
print("\n🔒 تم إغلاق الاتصال بقاعدة البيانات بنجاح!")
```
