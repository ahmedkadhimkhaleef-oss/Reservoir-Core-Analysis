```python
import numpy as np
import matplotlib.pyplot as plt

# 1. تعريف المتغيرات الأساسية للبئر (DCA Parameters)
q_i = 1200      # الإنتاج الابتدائي (برميل/يوم)
d = 0.005       # معدل الهبوط اليومي (0.5% يومياً)
days = np.arange(0, 365) # محاكاة لـ 365 يوماً

# 2. تطبيق معادلة Arps للهبوط الأسي (Exponential Decline)
q_t = q_i * np.exp(-d * days)

print("✅ تم تعريف دالة الهبوط الأسي بنجاح وجاهزة للرسم الهندسي!")
```

```python
# 1. تجهيز لوحة الرسم
plt.figure(figsize=(10, 6))

# 2. رسم منحنى هبوط الإنتاج الفعلي مع الزمن
plt.plot(days, q_t, color='green', linewidth=2.5, label='Oil Production Forecast ($q_t$)')

# 3. إضافة خط أفقياً يمثل الحد الاقتصادي الحرّج للبئر (Economic Limit)
economic_limit = 200
plt.axhline(y=economic_limit, color='red', linestyle='--', linewidth=2, 
            label=f'Economic Limit ({economic_limit} bbl/day)')

# 4. إضافة اللمسات الهندسية (العناوين والشبكة)
plt.title('Reservoir Decline Curve Analysis (DCA) - Exponential Model', fontsize=14, fontweight='bold')
plt.xlabel('Time (Days)', fontsize=12)
plt.ylabel('Oil Production Rate (bbl/day)', fontsize=12)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(fontsize=11)

# 5. عرض المنحنى
plt.show()
```

```python
# 1. تطبيق معادلة الـ EUR للهبوط الأسي
EUR = (q_i - economic_limit) / d

# 2. طباعة النتيجة بشكل احترافي
print("🏆 تقرير الاحتياطي الاستراتيجي للبئر:")
print("---------------------------------------")
print(f"🔹 إجمالي النفط القابل للاستخلاص (EUR): {EUR:,.0f} برميل")
print("---------------------------------------")
```
